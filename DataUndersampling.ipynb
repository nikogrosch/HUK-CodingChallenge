```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_pickle('Daten.pkl')
```

```python
df
```

## Mappen der Kategorischen Features in zahlen 

```python
#umwandeln von Strings in Zahlen 
df['Vorschaden'] = df['Vorschaden'].map({'No': 0, 'Yes': 1})
df['Geschlecht']= df['Geschlecht'].map({'Male': 0, 'Female': 1})
df['Alter_Fzg'] = df['Alter_Fzg'].map({'1-2 Year': 1, '< 1 Year': 0, '> 2 Years': 2})
#df = pd.get_dummies(df, columns=[]) ## falls Nominale Features in Zeilen umgewandelt werden sollen 
df
```

```python
nancount = df.isna().sum()
nancount
```

```python

df = df.drop(columns = ['id'])
```

## Korrelationen

```python
# Hier schaue ich mir die Korrelationen an bevor der Datensatz over oder undersampeld wird
korrelationen = df.corr()
#print(Korrelationen)

# Heatmap Korrelationsmatrix
plt.figure(figsize=(8, 6))
sns.heatmap(korrelationen, annot=True, cmap='hsv', fmt=".2f", linewidths=0.5)
plt.title('Korrelationsmatrix')
plt.show()
```

## Balancieren des DS durch undersampling

```python
pos = df[df['Interesse'] == 1]
neg = df[df['Interesse'] == 0]

Undersampling_neg = neg.sample(n=(len(pos)), random_state=42)
df = pd.concat([pos, Undersampling_neg])
df = df.sample(frac=1).reset_index(drop=True)
```

## Verteilungen nach Untersampling

```python
features = df.columns

# Erstelle ein kleines Diagramm für jedes Feature
plt.figure(figsize=(15, 10))

for i, feature in enumerate(features):
    plt.subplot(4, 4, i + 1) 
    plt.hist(df[feature], bins=20, edgecolor='black')
    plt.title(f"Verteilung von {feature}")
    plt.xlabel(feature)
    plt.ylabel('Häufigkeit')

plt.tight_layout()  # Optimiert das Layout der Diagramme
plt.show()
```

## Verhältnis klassen

```python
category_counts = df['Interesse'].value_counts()
fig, ax = plt.subplots(1, figsize=(10, 8))
bars = ax.bar(category_counts.index, category_counts, color= 'yellow')
ax.bar_label(bars, padding=5, fontsize=10, fontweight='bold')
plt.ylabel('Anzahl')
plt.xlabel('Interesse')
plt.show()
```

## Korrelation nach Undersampling

```python
#hier sind die Korrelationen nach dem Over oder Undersampling um sicherzugehen das der datensatz nicht übermäßig verzogen wird
korrelationen = df.corr()
#print(Korrelationen)

# Heatmap Korrelationsmatrix
plt.figure(figsize=(8, 6))
sns.heatmap(korrelationen, annot=True, cmap='hsv', fmt=".2f", linewidths=0.5)
plt.title('Korrelationsmatrix')
plt.show()
```

## Schauen ob Datensatz gemischt

```python
fig, (ax2) = plt.subplots(1, figsize = (16,8))
scatter = ax2.scatter(df.index, df['Interesse'], c='red', marker = 'x')
plt.ylabel('Interesse')
plt.show()
```
